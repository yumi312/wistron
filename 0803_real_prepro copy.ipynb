```python
def Feature_Engineering_all(x):
    df = {}
    # 在職月數、年數、個人月資
    df["LIFETIME_MONTH"] = raw_data_month.loc[x.iloc[-1].EMPLOYEE_ID].lifetime
    df["LIFETIME_YEAR"] = (raw_data_month.loc[x.iloc[-1].EMPLOYEE_ID].lifetime) // 12
    df["PERSON_TOTAL_MONTH"] = raw_data_month.loc[x.iloc[-1].EMPLOYEE_ID].EXPERIENCE_MONTH
    
    # 年齡、離職部門離職率、換過幾個部門、離職主管、換過幾個主管
    df["AGE"] = x.iloc[-1].AGE_LAST
    # df[""] # 離職部門的離職率
    df["CHG_NUM_DEP"] = len(x.DEPARTMENT_ID_LAST.unique())
    df["GRADE"] = x.iloc[-1].GRADE_NEW
    df["SUPERVISOR"] = x.iloc[-1].SUPERVISOR_LAST
    df["CHG_NUM_SUP"] = len(x.SUPERVISOR_LAST.unique())

    # 升過幾次等 
    df["GRADE_NUM"] = math.ceil(x.GRADE_DISTINCT_6M.sum() /6)

    # 離職時部門新人比例、部門周轉率
    df["RATIO_JUNIOR_LESS3"] = x.iloc[-1].RATIO_JUNIOR_LESS3
    df["RATIO_JUNIOR_LESS6"] = x.iloc[-1].RATIO_JUNIOR_LESS6
    df["RATIO_JUNIOR_LESS12"] = x.iloc[-1].RATIO_JUNIOR_LESS12
    df["RATIO_JUNIOR_LESS36"] = x.iloc[-1].RATIO_JUNIOR_LESS36
    df["DEPT_TURNOVER_RATIO"] = x.iloc[-1].DEPT_TURNOVER_RATIO

    # 員工之主管的分數
    if (len(x.tail(6)['SUPERVISOR_LAST'].unique())==1):
        df['SUPERVISOR_COR_SCORE'] = x.tail(6)['SUPERVISOR_COR_SCORE'].mean()## 有缺漏
        df['SUPERVISOR_PM2_SCORE'] = x.tail(6)['SUPERVISOR_PM2_SCORE'].mean()## 有缺漏
    else:
        df['SUPERVISOR_COR_SCORE'] = x[x['SUPERVISOR_LAST'] == x.tail(1)['SUPERVISOR_LAST'].unique().item()]['SUPERVISOR_COR_SCORE'].mean()## 有缺漏
        df['SUPERVISOR_PM2_SCORE'] = x[x['SUPERVISOR_LAST'] == x.tail(1)['SUPERVISOR_LAST'].unique().item()]['SUPERVISOR_PM2_SCORE'].mean()## 有缺漏
    
    # 在職期間的請假時數、加班時數、平日、假日、主管加班時數、平日、假日、部門加班時數(有問題，他換過部門?)、專案使用率、非專案使用率、會議時間、專案數、遲到次數(平均)
    df["ABS"] = x["TOTAL_ABS"].mean()
    df["OT"] = x["TOTAL_OT"].mean()
    df["REGULAR_OT"] = x["REGULAR_TOTAL_OT"].mean()
    df["HOLIDAY_OT"] = x["HOLIDAY_TOTAL_OT"].mean()

    df["SUPERVISOR_OT"] = x["SUPERVISOR_TOTAL_OT"].mean()
    df["SUPERVISOR_REGULAR_OT"] = x["SUPERVISOR_REGULAR_TOTAL_OT"].mean()
    df["SUPERVISOR_HOLIDAY_OT"] = x["SUPERVISOR_HOLIDAY_TOTAL_OT"].mean()

    df["DEP_OT"] = x["DEP_TOTAL_OT"].mean()

    df["PROJ_UTILIZATION"] = x["EMP_PROJ_UTILISATION"].mean()
    df["NONPROJ_UTILIZATION"] = x["RATIO_NONPROJ_HOURS_EMP"].mean()
    df["MEETING_HOURS"] = x["MEETING_HOURS_EMP"].mean()
    df["PROJ_NUM"] = x["NUM_PROJ_EMP"].mean()
    df["CLOCK_LATE_NUM"] = x["NUM_COUNT_CLOCK_LATE"].fillna(0).mean()

    # 3個月平均個人加班時數 > 3個月部門加班時數的平均次數
    df["IND_TOTAL_OT_3M"] = x["IND_TOTAL_OT_3M"].mean()
    df["IND_TOTAL_OT_6M"] = x["IND_TOTAL_OT_6M"].mean()
    
    # 核心職能、人員管理分數低
    df["COR_SCORE_LOW"] = x["COR_SCORE_LOW"].fillna(0).mean()
    df["PM2_SCORE_LOW"] = x["PM2_SCORE_LOW"].fillna(0).mean()
    
    # 在職期間升A、B等的次數
    df["PROMOTION_A_NUM"] = math.ceil(x["PROMOTION_TO_A_IN_2Y"].sum() / 12)
    df["PROMOTION_B_NUM"] = math.ceil(x["PROMOTION_TO_B_IN_5Y"].sum() / 60)

    return pd.Series(df)
```

```python
def Feature_Engineering_head(x):
    df = {}
    # 在職期間的請假時數、加班時數、平日、假日、主管加班時數、平日、假日、部門加班時數(有問題，他換過部門?)、專案使用率、非專案使用率、會議時間、專案數、遲到次數(平均)
    df["ABS_head"] = x["TOTAL_ABS"].mean()
    df["OT_head"] = x["TOTAL_OT"].mean()
    df["REGULAR_OT_head"] = x["REGULAR_TOTAL_OT"].mean()
    df["HOLIDAY_OT_head"] = x["HOLIDAY_TOTAL_OT"].mean()

    df["SUPERVISOR_OT_head"] = x["SUPERVISOR_TOTAL_OT"].mean()
    df["SUPERVISOR_REGULAR_OT_head"] = x["SUPERVISOR_REGULAR_TOTAL_OT"].mean()
    df["SUPERVISOR_HOLIDAY_OT_head"] = x["SUPERVISOR_HOLIDAY_TOTAL_OT"].mean()

    df["DEP_OT_head"] = x["DEP_TOTAL_OT"].mean()

    df["PROJ_UTILIZATION_head"] = x["EMP_PROJ_UTILISATION"].mean()
    df["NONPROJ_UTILIZATION_head"] = x["RATIO_NONPROJ_HOURS_EMP"].mean()
    df["MEETING_HOURS_head"] = x["MEETING_HOURS_EMP"].mean()
    df["PROJ_NUM_head"] = x["NUM_PROJ_EMP"].mean()
    df["CLOCK_LATE_NUM_head"] = x["NUM_COUNT_CLOCK_LATE"].fillna(0).mean()

    # 3個月平均個人加班時數 > 3個月部門加班時數的平均次數
    df["IND_TOTAL_OT_3M_head"] = x["IND_TOTAL_OT_3M"].mean()
    df["IND_TOTAL_OT_6M_head"] = x["IND_TOTAL_OT_6M"].mean()
    
    # 核心職能、人員管理分數低
    df["COR_SCORE_LOW_head"] = x["COR_SCORE_LOW"].fillna(0).mean()
    df["PM2_SCORE_LOW_head"] = x["PM2_SCORE_LOW"].fillna(0).mean()
    
    return pd.Series(df)
    
```
